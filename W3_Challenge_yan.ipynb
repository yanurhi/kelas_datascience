```python
import numpy as np

# Mencari mean menggunakan NumPy pada array atau kolom DataFrame
df_numpy = df_rating.loc[df_rating['Country Code'] == 208,'Aggregate rating'].values
mean_numpy = np.mean(df_numpy)
print(mean_numpy)

df_sorted = np.sort(df_numpy)
median_numpy = np.median(df_sorted)
print(median_numpy)

unique_values, counts = np.unique(df_numpy, return_counts=True)
mode = unique_values[np.argmax(counts)]
print(mode)
```

```python
df_numpy = df_rating.loc[df_rating['Country Code'] == 208,'Aggregate rating'].values
print("Range: ", np.ptp(df_numpy))  # ptp = "peak to peak"
print("Variance: ", np.var(df_numpy))
print("Standard Deviation: ", np.std(df_numpy))
```

```python
df_infer = df_rating[(df_rating['City'] == 'Ghaziabad') | (df_rating['City'] == 'Ankara')]
df_infer
```

one sided  
hipotesis penguji= resto ankara lebih enak dari ghazabad , apakah benar? tolong buktikan  
pakai convidence level 95%  

hipotesis penelitian  
H0 = ankara <= ghazabad  
H1 = ankara > ghazabad

```python
abad = df_infer[(df_infer['City'] == 'Ghaziabad')]
ankara = df_infer[(df_infer['City'] == 'Ankara')]
abad
```

```python
t_stat_one_sided, p_value_one_sided = stats.ttest_ind(ankara['Aggregate rating'], abad['Aggregate rating'], alternative='greater')
print("T-Statistic (One-sided):", t_stat_one_sided)
print("P-value (One-sided):", p_value_one_sided)
```

Jadi, berdasarkan nilai p-value sebesar 1.1047464363025233e-12, maka p value < 0,05 kita dapat menyimpulkan bahwa H0 ditolak sehingga resto ankara lebih enak dari ghazibad
